```python
import cv2
import os

def extract_frames(full_path, output_path, interval=1):  # interval = 1 to save one frame every second, int = 0.5 to save two frames, etc..
    # get the filename to save as specific folder
    file_name = os.path.basename(full_path).split('.')[0]

    # create specific output folder
    output_file = os.path.join(output_path, file_name)
    if os.path.exists(output_file):
        print(f"Skipping {file_name}, frames already extracted.")
        return
    os.makedirs(output_file, exist_ok=True)

    # read the video file
    cap = cv2.VideoCapture(full_path)
    if not cap.isOpened():
        print(f"Failed to open video file: {full_path}")
        return

    fps = cap.get(cv2.CAP_PROP_FPS)
    if fps <= 0:
        print(f"Invalid FPS value: {fps} for file: {full_path}")
        return

    print("Processing:", file_name, "with fps:", fps)
    count = 0
    frame_idx = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        if int(count % round(fps * interval)) == 0:
            frame_path = os.path.join(output_file, f"{file_name}_{frame_idx:04d}.jpg")
            cv2.imwrite(frame_path, frame)
            frame_idx += 1
        count += 1
    cap.release()

file_path = "C:/Users/yehte/Downloads/Ye Htet/Projects/TikTok/downloaded_videos"
output_path = "C:/Users/yehte/Downloads/Ye Htet/Projects/TikTok/zero_shot_face_recognition/third_discussion/extracted_frames"
for root, dirs, files in os.walk(file_path):
    for file in files:
        if file.lower().endswith((".mp4", ".avi", ".mov", ".mkv")):
            full_path = os.path.join(root, file)
            extract_frames(full_path, output_path)
```
